### Storytelling métier : pourquoi financer ce modèle ?

Aujourd’hui, le marché des ordinateurs portables est très difficile à lire pour un acheteur ou un revendeur. Deux ordinateurs peuvent sembler proches visuellement, mais avoir des prix très différents selon leur marque, leur processeur, leur RAM, leur stockage, leur carte graphique ou leur résolution d’écran.

Pour un client, cela peut entraîner une difficulté à savoir si un ordinateur est vendu à un prix juste. Pour un revendeur, cela peut rendre complexe la fixation d’un prix cohérent et compétitif.

Notre modèle répond à ce besoin en proposant une estimation automatique du prix d’un laptop à partir de ses caractéristiques techniques. L’objectif n’est pas de remplacer l’expertise humaine, mais de fournir un outil d’aide à la décision rapide, cohérent et basé sur les données.

### Intérêt stratégique du modèle

Le financement de ce modèle peut permettre de développer un outil d’aide à la décision utile pour plusieurs profils :

- les acheteurs, qui veulent vérifier si un prix est cohérent ;
- les revendeurs, qui souhaitent fixer des prix compétitifs ;
- les plateformes de vente, qui peuvent proposer une estimation automatique du prix ;
- les entreprises, qui doivent acheter ou renouveler un parc informatique.

Un tel outil peut réduire le temps d’analyse, limiter les erreurs d’évaluation et améliorer la transparence du marché.

```python
import plotly.express as px

price_by_company = (
    df
    .groupby("Company", as_index=False)["Price"]
    .mean()
    .sort_values(by="Price", ascending=False)
)

fig = px.bar(
    price_by_company,
    x="Company",
    y="Price",
    title="Prix moyen des laptops par marque",
    labels={
        "Company": "Marque",
        "Price": "Prix moyen"
    }
)

fig.show()
```

```python
import plotly.graph_objects as go

features = [
    "Ram",
    "Cpu_Frequence_GHz",
    "weight_kg",
    "Res_Width",
    "Res_Height",
    "Inches"
]

fig = go.Figure()

# Graphique initial
fig.add_trace(
    go.Scatter(
        x=df[features[0]],
        y=df["Price"],
        mode="markers",
        text=df["Company"],
        customdata=df[["Company", "TypeName", "Cpu_Gamme", "has_ssd"]],
        hovertemplate=
            "Valeur : %{x}<br>" +
            "Prix : %{y}<br>" +
            "Marque : %{customdata[0]}<br>" +
            "Type : %{customdata[1]}<br>" +
            "CPU : %{customdata[2]}<br>" +
            "SSD : %{customdata[3]}<extra></extra>"
    )
)

buttons = []

for feature in features:
    buttons.append(
        dict(
            label=feature,
            method="update",
            args=[
                {"x": [df[feature]]},
                {
                    "title": f"Prix des laptops selon {feature}",
                    "xaxis": {"title": feature},
                    "yaxis": {"title": "Prix"}
                }
            ]
        )
    )

fig.update_layout(
    title=f"Prix des laptops selon {features[0]}",
    xaxis_title=features[0],
    yaxis_title="Prix",
    updatemenus=[
        dict(
            buttons=buttons,
            direction="down",
            showactive=True,
            x=0.1,
            y=1.15
        )
    ]
)

fig.show()
```
