### Imputation des valeurs manquantes

Dans ce jeu de données, il n’y a pas de stratégie d’imputation complexe à mettre en place à ce stade. Nous commençons par supprimer les 30 lignes entièrement vides détectées lors de l’audit, puis nous vérifions les valeurs manquantes restantes pour décider du traitement le plus adapté.

La seule valeur manquante réellement traitée plus loin concerne la colonne Weight, pour laquelle une valeur aberrante est remplacée par la médiane après conversion en numérique. Cette approche limite la perte d’information tout en conservant une valeur représentative de la distribution.

```python
# Supresion des lignes nulles

print("Dimensions avant suppression des lignes nulles :")
print(df_raw.shape)

# Identifier les lignes entièrement nulles
df_null_rows = df_raw[df_raw.isnull().all(axis=1)]

print("\nNombre de lignes entièrement nulles détectées :")
print(len(df_null_rows))

print("\nAperçu des lignes entièrement nulles :")
display(df_null_rows)

# Supprimer les lignes entièrement nulles
df_without_null_rows = df_raw.dropna(how="all")

print("\nDimensions après suppression des lignes nulles :")
print(df_without_null_rows.shape)

print("\nNombre de valeurs manquantes restantes par colonne :")
display(df_without_null_rows.isnull().sum())
```

### Suppression des valeurs dupliquées

Une fois les lignes entièrement vides supprimées, nous vérifions la présence de doublons exacts dans le jeu de données. Cette étape permet d’éviter de conserver plusieurs fois la même observation et de fausser les statistiques descriptives ou les modèles d’analyse.

L’audit initial avait signalé 29 lignes dupliquées, mais elles correspondaient aux enregistrements entièrement vides. Après suppression de ces lignes, il ne reste plus de doublons à conserver, ce qui confirme que le nettoyage précédent a bien éliminé ces répétitions.


```python
# gestion des duplicates

print("Dimensions avant suppression des doublons :")
print(df_without_null_rows.shape)

# Identifier les lignes dupliquées
df_duplicates = df_without_null_rows[df_without_null_rows.duplicated()]

print("\nNombre de lignes dupliquées détectées :")
print(len(df_duplicates))

print("\nAperçu des lignes dupliquées :")
display(df_duplicates)

# Supprimer les doublons
df_clean = df_without_null_rows.drop_duplicates()

print("\nDimensions après suppression des doublons :")
print(df_clean.shape)

print("\nNombre de doublons restants :")
print(df_clean.duplicated().sum())


# constat : les lignes dupliquées correspondaient aux lignes nulles
```

### Nettoyage des séries

Cette partie transforme les variables textuelles les plus riches en variables plus exploitables pour l’analyse. L’idée est de conserver l’information utile tout en supprimant les redondances ou les formats difficiles à manipuler en l’état.

#### Nettoyage de la série ScreenResolution

La colonne ScreenResolution mélange plusieurs informations dans une seule chaîne de caractères. Pour la rendre exploitable, nous commençons par extraire la résolution numérique au format largeur × hauteur à l’aide d’une expression régulière.

Nous créons ensuite deux nouvelles colonnes, Res_Width et Res_Height, afin de séparer proprement ces dimensions. En parallèle, un indicateur binaire Is_IPS est ajouté pour signaler la présence de la technologie IPS dans l’écran.

Une fois ces informations extraites et vérifiées, la colonne d’origine ScreenResolution devient redondante et peut être supprimée.

```python
# Nettoyage de ScreenResolution
# Extraction de la résolution (ex: 2560x1600)
# On cherche le format XXXXxYYYY à la fin de la chaîne
resolution_regex = r'(\d+)x(\d+)'
df_clean[['Res_Width', 'Res_Height']] = df_clean['ScreenResolution'].str.extract(resolution_regex).astype(int)
df_clean['Is_IPS'] = df_clean['ScreenResolution'].str.contains('IPS').astype(int)

# Vérification de la transformation
print("Vérification de la transformation après nettoyage")
display(df_clean)

# Suppression de la colonne d'origine
df_clean = df_clean.drop(columns=["ScreenResolution"])
print("Affichage des données finales")
display(df_clean.head())

print("Affichage des séries")
display(df_clean.columns)
```

#### Nettoyage de la série CPU

La colonne Cpu combine plusieurs informations dans une seule chaîne, notamment la marque, la gamme et la fréquence. Pour simplifier l’analyse, nous isolons d’abord les trois premiers mots afin de récupérer la gamme principale du processeur.

Nous extrayons ensuite la fréquence exprimée en GHz à l’aide d’une expression régulière, puis nous la convertissons en variable numérique. Cela permet d’utiliser cette information dans des calculs ou des modèles sans traitement supplémentaire.

Une fois ces deux informations créées et contrôlées, la colonne Cpu d’origine devient redondante et peut être supprimée.

```python
# Nettoyage de Cpu
# --- TRAITEMENT DE LA COLONNE CPU ---

# On extrait les 3 premiers mots pour obtenir la marque et la gamme principale (ex: Intel Core i5)
df_clean["Cpu_Gamme"] = df_clean["Cpu"].str.split().str[0:3].str.join(" ")

# On isole la fréquence numérique juste avant "GHz" et on la convertit en nombre décimal (float)
df_clean["Cpu_Frequence_GHz"] = (
    df_clean["Cpu"].str.extract(r"(\d+\.?\d*)GHz$").astype(float)
)

# Vérification de l'extraction des informations CPU
print("Vérification de l'extraction des informations CPU")
display(df_clean[["Cpu", "Cpu_Gamme", "Cpu_Frequence_GHz"]].head(10))

# Suppression de la colonne Cpu d'origine devenue redondante
df_clean = df_clean.drop(columns=["Cpu"])


# # --- TRAITEMENT DE LA COLONNE GPU ---

# # On récupère uniquement le premier mot de la chaîne (le constructeur)
# df_clean["Gpu_Constructeur"] = df_clean["Gpu"].str.split().str[0]

# display(df_clean[["Cpu", "Cpu_Gamme", "Cpu_Frequence_GHz", "Gpu", "Gpu_Constructeur"]].head())
```

#### Nettoyage de la série Memory

La colonne Memory est plus complexe car elle mélange plusieurs informations dans une seule chaîne, avec des capacités en GB et en TB ainsi que plusieurs types de stockage. Nous commençons par observer les valeurs les plus fréquentes pour comprendre sa structure.

Ensuite, nous convertissons toutes les capacités en gigaoctets afin d’unifier les unités. En parallèle, nous créons des indicateurs binaires pour signaler la présence de SSD, HDD, Flash Storage ou Hybrid.

Enfin, une colonne total_memory_gb est calculée pour résumer la capacité totale de stockage, puis la colonne Memory d’origine est supprimée car elle est devenue trop redondante.

```python
# Nettoyage de Memory
# observation de la colonne
display(df_clean["Memory"].value_counts().head(30))
# La colonne memory contient plusieurs informations mélangées dans du texte, ces informations ne sont pas exploitables, il faut diviser cette colonne en plusieurs. Il ya plusieurs types de mémoires et des unités non identiques : TB, GB.

# Conversion des GB/TB en GB

import re

def convert_memory_to_gb(memory):
    memory = str(memory)
    total_gb = 0

    # chercher toutes les valeurs du type 128GB, 1TB, 512GB, etc.
    matches = re.findall(r"(\d+\.?\d*)\s*(GB|TB)", memory)

    for value, unit in matches:
        value = float(value)

        if unit == "TB":
            value = value * 1024

        total_gb += value

    return total_gb

# Extraction des types de mémoire
df_clean["has_ssd"] = df_clean["Memory"].str.contains("SSD", case=False, na=False).astype(int)
df_clean["has_hdd"] = df_clean["Memory"].str.contains("HDD", case=False, na=False).astype(int)
df_clean["has_flash"] = df_clean["Memory"].str.contains("Flash Storage", case=False, na=False).astype(int)
df_clean["has_hybrid"] = df_clean["Memory"].str.contains("Hybrid", case=False, na=False).astype(int)


# Extraction de la capacité totale en GB
df_clean["total_memory_gb"] = df_clean["Memory"].apply(convert_memory_to_gb)

# Vérification de la transformation
print("Vérification de la transformation")
display(df_clean[[
    "Memory",
    "has_ssd",
    "has_hdd",
    "has_flash",
    "has_hybrid",
    "total_memory_gb"
]].head(20))

# Suppression de la colonne d'origine
df_clean = df_clean.drop(columns=["Memory"])
display(df_clean.head())
print("Affichage des séries")
print(df_clean.columns)
```

#### Nettoyage de la série GPU

La colonne GPU contient à la fois le constructeur et des informations de gamme. Nous commençons ici par isoler la marque principale afin de créer des variables simples à interpréter pour la suite de l’analyse.

À partir de cette marque, nous créons ensuite des indicateurs binaires pour distinguer les puces Intel, Nvidia et AMD. Cette étape facilite les comparaisons entre machines et simplifie l’usage de la variable dans les analyses futures.

Une fois ces variables créées et vérifiées, la colonne GPU d’origine peut être supprimée.

```python
# Nettoyage de Gpu
# Observation des données de départ 
display(df_clean["Gpu"].value_counts().head(30))

# cette colonne contient de nombreuses informations : marque, gamme, modèle précis. POur des raisons de simplicité nous faisons le choix d'extraire uniquement la marque pour commencer.

# Extraction de la marque du GPU
def extract_gpu_brand(gpu):
    gpu = str(gpu).lower()

    if gpu.startswith("intel"):
        return "Intel"
    elif gpu.startswith("nvidia"):
        return "Nvidia"
    elif gpu.startswith("amd"):
        return "AMD"
    else:
        return "Other"

df_clean["gpu_brand"] = df_clean["Gpu"].apply(extract_gpu_brand)

# Création de colonnes binaires
df_clean["has_intel_gpu"] = (df_clean["gpu_brand"] == "Intel").astype(int)
df_clean["has_nvidia_gpu"] = (df_clean["gpu_brand"] == "Nvidia").astype(int)
df_clean["has_amd_gpu"] = (df_clean["gpu_brand"] == "AMD").astype(int)

# Vérification
display(df_clean[[
    "Gpu",
    "gpu_brand",
    "has_intel_gpu",
    "has_nvidia_gpu",
    "has_amd_gpu"
]].head(20))

display(df_clean["gpu_brand"].value_counts())

# Suppression de la colonne d'origine
df_clean = df_clean.drop(columns=["Gpu"])
display(df_clean.head())
```

#### Nettoyage de la série Weight

La colonne Weight est initialement stockée sous forme de texte avec l’unité kg. Nous la nettoyons en retirant cette unité puis en convertissant la valeur en numérique afin de pouvoir l’exploiter plus facilement.

Pendant cette étape, nous repérons aussi les valeurs qui ne peuvent pas être converties correctement, comme le caractère ? présent dans certaines lignes. Après remplacement de cette anomalie par une valeur manquante, nous imputons cette dernière avec la médiane.

Cette stratégie conserve l’information tout en évitant de supprimer une ligne entière pour une seule valeur défectueuse. La colonne d’origine peut ensuite être retirée au profit de weight_kg.

```python
# Nettoyage Weight
# ici il s'agit de nettoyer une colonne texte en colonne numérique
# Observation des valeurs de départ
display(df_clean["Weight"].value_counts().head(30))

# Nettoyage temporaire de Weight pour identifier les problèmes
weight_cleaned_temp = (
    df_clean["Weight"]
    .astype(str)
    .str.replace("kg", "", regex=False)
    .str.strip()
)

# Conversion en numérique, les erreurs deviennent NaN
weight_numeric_temp = pd.to_numeric(weight_cleaned_temp, errors="coerce")

# Afficher les valeurs qui n'ont pas pu être converties
invalid_weights = df_clean[weight_numeric_temp.isna()][["Weight"]]

print("Nombre de valeurs Weight non convertibles :")
print(len(invalid_weights))

display(invalid_weights)

# une valeur s'affiche avec un ? je décide de remplacer la valeur manquante par la médiane
# je remplace le ? par NaN
df_clean["Weight"] = df_clean["Weight"].replace("?", np.nan)

# Créer une colonne numérique sans "kg"
df_clean["weight_kg"] = (
    df_clean["Weight"]
    .astype(str)
    .str.replace("kg", "", regex=False)
    .str.strip()
)

# Convertir en numérique
df_clean["weight_kg"] = pd.to_numeric(df_clean["weight_kg"], errors="coerce")

print("Nombre de valeurs manquantes après conversion :")
print(df_clean["weight_kg"].isnull().sum())

# Remplacer la valeur manquante par la médiane
median_weight = df_clean["weight_kg"].median()
df_clean["weight_kg"] = df_clean["weight_kg"].fillna(median_weight)

print("Poids médian utilisé pour remplacer la valeur manquante :")
print(median_weight)

print("Nombre de valeurs manquantes après remplacement :")
print(df_clean["weight_kg"].isnull().sum())

# Vérification
display(df_clean[["Weight", "weight_kg"]].iloc[200:215])

# Suppression de la colonne d'origine
df_final = df_clean.drop(columns=["Weight"])
display(df_clean.head())
```

### Nettoyage des valeurs "?"

Lors des étapes suivantes, une valeur non numérique est retrouvée, il s'agit d'un "?". Nous nettoyons cette valeur dans cette étape.

```python
# Vérification du nombre de valeurs "?" par colonne
question_marks = (df_clean == "?").sum()
print("Valeurs '?' par colonne : ")
print(question_marks[question_marks > 0])

# Remplacement des "?" par des valeurs manquantes NaN
df_clean = df_clean.replace("?", np.nan)

# Vérification des valeurs manquantes après remplacement
print("Valeurs manquantes après remplacement : ")
print(df_clean.isna().sum())

# Suppression des lignes contenant des valeurs manquantes
df_final = df_clean.dropna()

# Vérification finale
print("Vérification finale : ")
print(df_final.isna().sum())
```
